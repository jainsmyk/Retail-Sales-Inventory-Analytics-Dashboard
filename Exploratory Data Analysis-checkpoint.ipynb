```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings("ignore") 
df= pd.read_csv(r"C:\Users\hp\Desktop\Zepto Project\Zepto_v4.csv")
df.head()
print(df.info())
print(df.describe())
print(df['category'].value_counts())

df["mrp"] = df["mrp"] / 100
df["discountedSellingPrice"] = df["discountedSellingPrice"] / 100
df["discountPercent"] = ((df["mrp"] - df["discountedSellingPrice"]) / df["mrp"] * 100).round(2)


```

```python
print(df.info())
print(df.describe())
print(df['category'].value_counts())
```

```python
df.shape
```

```python
###Top 10 best value products based on discount percentage###
```

```python
top_10_value = df.sort_values(by="discountPercent", ascending=False).head(10)
print(top_10_value[["category", "name", "mrp", "discountPercent", "discountedSellingPrice"]])

plt.figure(figsize=(10,6))
sns.barplot(
    data=top_10_value,
    x="discountPercent",
    y="name",
    palette="viridis"
)
plt.title("Top 10 Best Value Products (Highest Discount %)", fontsize=14)
plt.xlabel("Discount Percentage")
plt.ylabel("Product Name")
plt.show()
```

```python
### Convert products in to Categories like Low,Medium,Bulk ###
```

```python
def categorize_mrp(mrp):
    if mrp < 100:
        return "Low"
    elif mrp < 300:
        return "Medium"
    else:
        return "Bulk"

df["MRP_Category"] = df["mrp"].apply(categorize_mrp)
print(df["MRP_Category"].value_counts())

plt.figure(figsize=(8,4))
sns.countplot(x="MRP_Category", data=df, palette="Set2")
plt.title("Product Categories by MRP")
plt.xlabel("Category")
plt.ylabel("Count of Products")
plt.show()
```

```python
### Top 10 Low Discount Products ###
```

```python
low_discount_products = (
    df.sort_values(by="discountPercent", ascending=True)
      .head(10)[["category", "name", "mrp", "discountPercent", "discountedSellingPrice"]]
)

print("Top 10 Products with Very Low Discount:\n", low_discount_products)
plt.figure(figsize=(12,6))
sns.barplot(
    x=low_discount_products["discountPercent"],
    y=low_discount_products["name"],
    palette="viridis"
)
plt.xlabel("Discount (%)")
plt.ylabel("Product Name")
plt.title("Top 10 Products with Very Low Discount")
plt.show()
```
